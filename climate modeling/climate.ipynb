# Climate Change Modeling


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
```

# Set random seed for reproducibility

```python
np.random.seed(42)
```

# Generate synthetic temperature anomaly data (1850-2025)

```python
years = np.arange(1850, 2026)
n_years = len(years)
```

# Simulate a linear warming trend with noise

```python
trend = 0.01 * (years - 1850)  # 0.01Â°C per year warming
noise = np.random.normal(0, 0.2, n_years)  # Random noise
temp_anomaly = trend + noise
```

# Create DataFrame


```python
df = pd.DataFrame({'Year': years, 'Temp_Anomaly': temp_anomaly})
```

# Calculate linear regression


```python
slope, intercept, r_value, p_value, std_err = stats.linregress(df['Year'], df['Temp_Anomaly'])
line = intercept + slope * df['Year']
```

# Plotting


```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Year'], df['Temp_Anomaly'], color='blue', alpha=0.5, label='Temperature Anomaly')
plt.plot(df['Year'], line, color='red', label=f'Trend (slope: {slope:.4f}Â°C/year)')
plt.xlabel('Year')
plt.ylabel('Temperature Anomaly (Â°C)')
plt.title('Global Temperature Anomaly (1850-2025)')
plt.legend()
plt.grid(True)
plt.show()
```

# Calculate and print trend statistics


```python
print(f"Trend Slope: {slope:.4f}Â°C/year")
print(f"R-squared: {r_value**2:.4f}")
print(f"P-value: {p_value:.4e}")
```

# Simple future projection (next 50 years)


```python
future_years = np.arange(2026, 2076)
future_anomaly = intercept + slope * future_years
future_df = pd.DataFrame({'Year': future_years, 'Temp_Anomaly': future_anomaly})
plt.figure(figsize=(10, 6))
plt.plot(df['Year'], df['Temp_Anomaly'], color='blue', label='Historical Data')
```

# Plot with projection


```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Year'], df['Temp_Anomaly'], color='blue', alpha=0.5, label='Historical')
plt.plot(df['Year'], line, color='red', label='Historical Trend')
plt.plot(future_df['Year'], future_df['Temp_Anomaly'], color='green', linestyle='--', label='Projected Trend')
plt.xlabel('Year')
plt.ylabel('Temperature Anomaly (Â°C)')
plt.title('Temperature Anomaly with 50-Year Projection')
plt.legend()
plt.grid(True)
plt.show()
```
